```python
import requests
import time
import csv
from datetime import datetime, timedelta

API_KEY = "8b8d744826de418b8ad03cca5eacc618"
URL = "https://newsapi.org/v2/everything"

queries = [
    "artificial intelligence",
    "AI jobs",
    "automation",
    "future of work",
    "AI impact on employment",
    "AI replacing humans",
    "machine learning jobs",
    "AI in workplace",
    "robots replacing humans",
    "automation impact on employment"
]


from_date = (datetime.utcnow() - timedelta(days=30)).strftime("%Y-%m-%d")

output_file = "ai_vs_human_news.csv"

# Colonnes du CSV
fieldnames = [
    "source",
    "title",
    "description",
    "content",
    "author",
    "published_at",
    "url",
    "keywords",
    "ingested_at"
]

all_articles_count = 0

with open(output_file, mode="w", newline="", encoding="utf-8") as file:
    writer = csv.DictWriter(file, fieldnames=fieldnames)
    writer.writeheader()

    for query in queries:
        for page in range(1, 6):  # pagination
            params = {
                "q": query,
                "language": "en",
                "from": from_date,
                "pageSize": 100,
                "page": page,
                "apiKey": API_KEY
            }

            response = requests.get(URL, params=params)
            data = response.json()
            articles = data.get("articles", [])

            print(f"ðŸ”Ž '{query}' | page {page} â†’ {len(articles)} articles")

            if not articles:
                break

            for article in articles:
                writer.writerow({
                    "source": "newsapi",
                    "title": article.get("title"),
                    "description": article.get("description"),
                    "content": article.get("content"),
                    "author": article.get("author"),
                    "published_at": article.get("publishedAt"),
                    "url": article.get("url"),
                    "keywords": "AI,automation,jobs",
                    "ingested_at": datetime.utcnow().isoformat()
                })
                all_articles_count += 1

                time.sleep(0.5)  # simulation streaming

print(f"\nâœ… Fichier CSV crÃ©Ã© : {output_file}")
print(f"ðŸ“Š Total articles enregistrÃ©s : {all_articles_count}")
```

```python
import csv

input_file = "ai_vs_human_news.csv"       # CSV original
output_file = "ai_vs_human_news_clean.csv"  # CSV nettoyÃ©

# Dictionnaire pour dÃ©tecter les doublons
unique_articles = {}

# Lecture du CSV original
with open(input_file, mode="r", encoding="utf-8") as infile:
    reader = csv.DictReader(infile)
    for row in reader:
        url = row.get("url")
        if url and url not in unique_articles:
            unique_articles[url] = row  # On garde seulement le premier article avec cette URL

# Ã‰criture du CSV nettoyÃ©
with open(output_file, mode="w", newline="", encoding="utf-8") as outfile:
    fieldnames = reader.fieldnames
    writer = csv.DictWriter(outfile, fieldnames=fieldnames)
    writer.writeheader()

    for article in unique_articles.values():
        writer.writerow(article)

print(f"\nâœ… Nettoyage terminÃ© !")
print(f"Nombre d'articles uniques : {len(unique_articles)}")
print(f"Fichier CSV final : {output_file}")
```
